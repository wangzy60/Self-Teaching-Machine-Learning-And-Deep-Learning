### torch.nn.functional

#### torch.nn.Module和torch.nn.Conv1d/2d/……这些已经实现好了的模型框架用起来非常全面，可以自动进行参数的更新等流程，但是它的初始化和使用相对复杂，如果有的时候，我们并不需要让参数进行自动更新，只是单纯的要使用一下conv或者pooling方法，应该怎么办呢，有没有更快速和简单就能使用这些网络层的方式呢？
#### 这就是torch.nn.functional的意义
#### torch.nn.functional下封装好了很多函数，基本攘括了torch.nn.Module下已经实现了的那些网络层，像conv，pooling，padding等都以函数的形式封装好了。这些函数通过提供输入数据和对应的层参数，例如卷积核shape之类的，可以进行一次正向传播，返回计算的结果。
#### 具体函数可以参照上一篇列出的各种层，每个层都包含自己的函数
