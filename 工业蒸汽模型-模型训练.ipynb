```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

#### 读取数据,检查数据是否正常


```python
train_data = pd.read_csv('zhengqi_train_change_feat.txt',sep='\t')
test_data = pd.read_csv('zhengqi_test_change_feat.txt',sep='\t')
train_data.info()
train_data
```

```python
test_data.info()
test_data
```

#### 将数据分成训练集和验证集

```python
from sklearn.model_selection import train_test_split
train_data_tmp,valid_data = train_test_split(train_data)

feature_list = [feat for feat in train_data.columns if feat not in ['target']]

train_feat = train_data_tmp[feature_list]
train_target = train_data_tmp['target']
valid_feat = valid_data[feature_list]
valid_target = valid_data['target']

train_feat.info()
print(train_target,'\n')

valid_feat.info()
print(valid_target)
```

```python
#线性回归模型
linear_model = LinearRegression()
linear_model.fit(train_feat,train_target)

linear_train_predic = linear_model.predict(train_feat)
linear_valid_predic = linear_model.predict(valid_feat)

print('linear_train_MSE:',mean_squared_error(linear_train_predic,train_target))
print('linear_valid_MSE:',mean_squared_error(linear_valid_predic,valid_target))
```

```python
#KNN临近模型
from sklearn.neighbors import KNeighborsRegressor

knn_model = KNeighborsRegressor(n_neighbors=8)
knn_model.fit(train_feat,train_target)

knn_train_predic = knn_model.predict(train_feat)
knn_valid_predic = knn_model.predict(valid_feat)

print('knn_train_MSE:',mean_squared_error(knn_train_predic,train_target))
print('knn_valid_MSE:',mean_squared_error(knn_valid_predic,valid_target))
```

```python
#决策树回归
from sklearn.tree import DecisionTreeRegressor

decision_tree_model = DecisionTreeRegressor()
decision_tree_model.fit(train_feat,train_target)

dt_train_predic = decision_tree_model.predict(train_feat)
dt_valid_predic = decision_tree_model.predict(valid_feat)

print('dt_train_MSE:',mean_squared_error(dt_train_predic,train_target))
print('dt_valid_MSE:',mean_squared_error(dt_valid_predic,valid_target))
```

```python
#随机森林回归
from sklearn.ensemble import RandomForestRegressor

random_forest_model = RandomForestRegressor(n_estimators=200)
random_forest_model.fit(train_feat,train_target)

rf_train_predic = random_forest_model.predict(train_feat)
rf_valid_predic = random_forest_model.predict(valid_feat)

print('rf_train_MSE:',mean_squared_error(rf_train_predic,train_target))
print('rf_valid_MSE:',mean_squared_error(rf_valid_predic,valid_target))
```

```python
#LGB回归
from lightgbm import LGBMRegressor

lgb_model = LGBMRegressor(learning_rate=0.01,
                         max_depth=-1,
                         n_estimators=5000,
                         boosting_type='gbdt',
                         random_state=2019,
                         objective='regression')

lgb_model.fit(X=train_feat,y=train_target,eval_metric='MSE',verbose=50)

lgb_train_predic = lgb_model.predict(train_feat)
lgb_valid_predic = lgb_model.predict(valid_feat)

print('lgb_train_MSE:',mean_squared_error(lgb_train_predic,train_target))
print('lgb_valid_MSE:',mean_squared_error(lgb_valid_predic,valid_target))
```

#### 用最好的模型对test数据进行预测，将结果写入txt文件

```python
best_model=lgb_model
test_predic = best_model.predict(test_data)
print(test_predic)
np.savetxt('predictions.txt',test_predic,delimiter='\t')
```
