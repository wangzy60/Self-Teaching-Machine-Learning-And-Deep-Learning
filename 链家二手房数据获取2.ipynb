```python
def sumCsvfile(csvfileNameList,houseDataHeader,sumCsvfileName):
    print('开始汇总文件')
    houseData = []
    for csvfileName in csvfileNameList:
        with open(csvfileName,newline='',encoding='utf-8') as csvfile:
            reader= csv.DictReader(csvfile)
            for row in reader:
                houseData.append(row) 
    with open(sumCsvfileName,'w',newline='',encoding='utf-8') as sumCsvfile:
        writer = csv.DictWriter(sumCsvfile,fieldnames=houseDataHeader,restval='暂无数据',extrasaction='ignore')
        writer.writeheader()
        for i in houseData:
            writer.writerow(i)
        print(f'文件汇总完成，文件名称为{sumCsvfileName},不算表头共{len(houseData)}行数据')
```
